```julia
supp1 = displacementConstraint("bottom", uy=0)
supp2 = displacementConstraint("left", ux=0)
force = load("top", fy=1);
```

```julia
f = loadVector(problem, [force])
showDoFResults(f, :vector, name="load")
```

```julia
q = solveDisplacement(problem, [force], [supp1, supp2])
S = solveStress(q);
```
