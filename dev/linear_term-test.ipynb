```julia
mat = Material("surface")
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f)
```

```julia
h = scalarField(Pu, "surface", 1)
M = [h 0; 0 h]
```

```julia
μ = mat.μ
λ = mat.λ

K = ∫(ε(Pu) ⋅ M ⋅ ε(Pu) * 2μ + Div(Pu) ⋅ Div(Pu) * λ)
```
