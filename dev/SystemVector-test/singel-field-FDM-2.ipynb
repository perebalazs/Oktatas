```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()

mat = Material("body")
PT = Problem([mat], type=:ScalarField, dim=2, field=:T, rhs_field=:q)
```

```julia
k = mat.k
ρ = mat.ρ
c = mat.c
```

```julia
K = ∫(Grad(PT) ⋅ k ⋅ Grad(PT))
C = ∫(PT ⋅ PT * (ρ * c))
```

```julia
q = ∫(PT ⋅ 1, Γ="right")
```

```julia
K2 = SystemMatrix([K;;])
q2 = SystemVector([q])
```

```julia
bc = BoundaryCondition("left", problem=PT, T=0)
```

```julia
T = solveField(K2, q2, support=[bc])
T = T
```

```julia
showDoFResults(q, name="q")
showDoFResults(T, name="T", visible=true)
```
