```julia
using LowLevelFEM
```

```julia
structured_box_mesh()

mat = Material("volume")
PT = Problem([mat], type=:TensorField, field=:T, rhs_field=:f)
```

```julia
K = вҲ«(TensorDiv(PT) вӢ… TensorDiv(PT))
K[:, :]
```

```julia
ff = VectorField(PT, "volume", [(x, y, z) -> x, (x, y, z) -> y^2, (x, y, z) -> z * y])
f = вҲ«(TensorDiv(PT) вӢ… ff, О©="volume")
#f = вҲ«(TensorDiv(PT) вӢ… [ff[1]; ff[2]; ff[3];;], О©="volume")
#f = вҲ«(TensorDiv(PT) вӢ… [1, 0, 0], О©="volume")
#f = вҲ«(TensorDiv(PT) вӢ… [ff[1], ff[2], 0], О©="volume")
```

```julia
bc = BoundaryCondition("left", Tx=0)
```

```julia
T = solveField(K, f, support=[bc])
```

```julia
using LinearAlgebra
err = вҲҮ вӢ… T - ff
norm(DoFs(elementsToNodes(err))[:])
```
