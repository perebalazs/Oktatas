```julia
using LowLevelFEM
```

```julia
using LinearAlgebra
```

```julia
structured_box_mesh()

openPreProcessor()

mat = Material("body")
Pu1 = Problem([mat], type=:Solid)
Pu2 = Problem([mat], type=:VectorField, dim=3)
```

```julia
K1 = stiffnessMatrix(Pu1)
K1[:, :]
```

```julia
μ = mat.μ
λ = mat.λ

K2 = 1 * ∫(SymGrad(Pu2) ⋅ (Diagonal([2, 2, 2, 1, 1, 1]) * μ) ⋅ SymGrad(Pu2)) + ∫(Div(Pu2) ⋅ λ ⋅ Div(Pu2))
K3 = 1 * symmetricGradientMatrix(Pu2, coefficient=μ) + gradDivMatrix(Pu2, coefficient=λ)
K2[:, :]
```

```julia
norm(K3[:, :] - K2[:, :])
```
