```julia
using LowLevelFEM
```

```julia
nel = 50
A = 1
E = 1.0
ρ = 1.0

n = 1000

model = line_mesh(n=nel)

mat = Material("body", E=E, ρ=ρ)

prob = Problem([mat], type=:Truss)
```

```julia

K = stiffnessMatrix(prob)
M = massMatrix(prob)

bc = BoundaryCondition("left", ux=0)
bc0 = BoundaryCondition("body", uy=0, uz=0)

Tmin = smallestPeriodTime(K, M, support=[bc, bc0])
dt = Tmin / π
tmax = n * dt

g0 = ScalarField(prob, "left", (x, y, z, t) -> 0.1 * sin(10t), steps=n, tmax=tmax)
h0 = ScalarField(prob, "right", (x, y, z, t) -> 0.5 * sin(5t), steps=n, tmax=tmax)
o = ScalarField(prob, "left", 0, steps=n, tmax=tmax)
g = VectorField([g0, o, o])
o = ScalarField(prob, "right", 0, steps=n, tmax=tmax)
h = VectorField([h0, o, o])

ld = LoadCondition("right", fx=h0)
f = loadVector(prob, [ld])

bc = BoundaryCondition("left", ux=g0)

u0 = vectorField(prob, "body", [0, 0, 0])
v0 = vectorField(prob, "body", [0, 0, 0])

u1, v1 = CDM(K, M, f, u0, v0, n, dt, support=[bc, bc0])
u2, v2 = HHT(K, M, f, u0, v0, n, dt, support=[bc, bc0])
```

```julia
showDoFResults(g, name="g")
u01 = showDoFResults(u1, :x)
u02 = showDoFResults(u2, :x)
plotOnPath("body", u01)
plotOnPath("body", u02)
```
