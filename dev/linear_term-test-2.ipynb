```julia
using LowLevelFEM, LinearAlgebra
```

```julia
gmsh.initialize()

structured_rect_mesh()
```

```julia
mat = Material("surface")
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f)
```

```julia
M = [2 0; 0 2]

f = [1, 2]
f = VectorField(Pu, "surface", [1, 2, 3])
```

```julia
A = ∫(Pu ⋅ f, Γ="top")
norm(A.a)
```

```julia
A = ∫(Pu ⋅ M ⋅ f, Γ="top")
norm(A.a)
```

```julia
A = ∫(Pu ⋅ M ⋅ M ⋅ f)
norm(A.a)
```
