```julia
using LowLevelFEM
```

```julia
structured_box_mesh()

mat = Material("volume", E=1)
Pu = Problem([mat], type=:VectorField, dim=3, field=:u, rhs_field=:f);
```

```julia
supp1 = BoundaryCondition("left", ux=0)
supp2 = BoundaryCondition("bottom", uy=0)
supp3 = BoundaryCondition("rear", uz=0)
```

```julia
K = вҲ«(Оө(Pu) вӢ… Оө(Pu) * 2mat.Ој + Div(Pu) вӢ… Div(Pu) * mat.О»)
K[:, :]
```

```julia
g = VectorField(Pu, "volume", [1, 0, 0])
h = scalarField(Pu, "volume", 1)
```

```julia
f = вҲ«(Pu вӢ… [1 0 0; 0 h 0; 0 0 1] вӢ… g, О“="right")
```

```julia
#f = вҲ«(Pu вӢ… [1, 0, 0], О“="right")
```

```julia
u = solveField(K, f, support=[supp1, supp2, supp3])
```
