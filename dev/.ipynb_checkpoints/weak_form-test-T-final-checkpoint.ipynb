```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()
# openGeometry("name.geo)
mat = Material("surface")
PT = Problem([mat], type=:ScalarField, dim=2, field=:T, rhs_field=:q)
```

```julia
suppL = BoundaryCondition("left", T=0)
suppR = BoundaryCondition("right", T=100)
```

```julia
K = вҲ«(Grad(PT) вӢ… Grad(PT) * mat.k)
K[:, :]
```

```julia
h = scalarField(PT, "surface", 1)
```

```julia
Kh = вҲ«(PT вӢ… [1;;] вӢ… h, О“="right")
```

```julia
Kh = вҲ«(PT вӢ… h, О“="right")
```

```julia
Kh = вҲ«(PT вӢ… [1;;] вӢ… 1, О“="right")
```

```julia
Kh = вҲ«(PT вӢ… [1;;] вӢ… 1, О“="right")
```

```julia
Kh = вҲ«(PT вӢ… [1;;] вӢ… PT, О“="right")
```

```julia
Kh = вҲ«(PT вӢ… [1;;] вӢ… Id(PT), О“="right")
```

```julia
Kh = вҲ«(PT вӢ… PT * 10000, О“="right")
Kh[:, :]
```

```julia
f = вҲ«(PT вӢ… 1, О“="right")
```

```julia
fh = вҲ«(PT вӢ… (10000 * 100), О“="right")
```

```julia
T = solveField(K + Kh, fh, support=[suppL])
```

```julia
showDoFResults(T, visible=false)
showDoFResults(-mat.k * вҲӮx(T), visible=true)
```
