```julia
csavar = Material("csavar", E=2e5, ν=0.3)
tipli = Material("tipli", E=100, ν=0.4)

prob = Problem([csavar, tipli])
```

```julia
supp = BoundaryCondition("fix", ux=0, uy=0, uz=0)
ld = load("load", fz=1)
```

```julia
u = solveDisplacement(prob, load=[ld], support=[supp])
showDoFResults(u, name="u", factor=1)
```

```julia
S = solveStress(u)
showElementResults(S, name="σ")
```

```julia
integrate(prob, "load", (x, y, z) -> 1)
```
