```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("gomb.geo")

mat = material("volu")
prob = Problem([mat])
```

```julia
function VectorField0(problem::Problem, phName::String, v::Vector; steps=1, tmin=0.0, tmax=tmin + (steps - 1))
    s1 = ScalarField(problem, phName, v[1], steps=steps, tmin=tmin, tmax=tmax)
    s2 = ScalarField(problem, phName, v[2], steps=steps, tmin=tmin, tmax=tmax)
    s3 = ScalarField(problem, phName, v[3], steps=steps, tmin=tmin, tmax=tmax)
    return VectorField([s1, s2, s3])
end
```

```julia
g = 9.81
v0x = 10
v0y = 20
tmax = 5
steps = 200
a = VectorField0(prob, "volu", [0, -g, 0], tmax=tmax, steps=steps)
v0 = VectorField0(prob, "volu", [v0x, v0y, 0], tmax=tmax, steps=steps)

x0 = ScalarField(prob, "volu", (x, y, z) -> x, tmax=tmax, steps=steps)
y0 = ScalarField(prob, "volu", (x, y, z) -> y, tmax=tmax, steps=steps)
z0 = ScalarField(prob, "volu", (x, y, z) -> z, tmax=tmax, steps=steps)

r0 = VectorField([x0, y0, z0]);
```

```julia
v = v0 + integrate(a)
r = r0 + integrate(v);
```

```julia
showDoFResults(a, name="a")
showDoFResults(v, name="v")
showDeformationResults(r, name="r")

openPostProcessor()
```

```julia
gmsh.finalize()
```
