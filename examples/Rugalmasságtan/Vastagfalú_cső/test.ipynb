```julia
using LowLevelFEM
```

```julia
structured_rect_mesh(x0=20, lx=20, ly=10, n=10, order=1)

mat = Material("body")
prob = Problem([mat], type=:AxiSymmetric)
```

```julia
bc1 = displacementConstraint("bottom", uy=0)
bc2 = displacementConstraint("top", uy=0)

ld = load("left", fx=1)
```

```julia
u = solveDisplacement(prob, support=[bc1, bc2], load=[ld])
```

```julia
A = solveStrain(u)
S = solveStress(u)
```

```julia
showDoFResults(u, name="u", visible=true)
showStrainResults(A, :x, name="╬╡R")
showStrainResults(A, :z, name="╬╡z")

sr = showStressResults(S, :x, name="╧ГR")
s╧Ж = showStressResults(S, :y, name="╧Гz")
sz = showElementResults(S, :z, name="╧Г╧Ж")
```

```julia
Rk = 40
Rb = 20
pb = 1
A = Rb^2 * Rk^2 * (-pb) / (Rk^2-Rb^2)
B = pb*Rb^2 / (Rk^2-Rb^2)
Sr = ScalarField(prob, "body", (x, y, z)->A/x^2+B)
S╧Ж = ScalarField(prob, "body", (x, y, z)->-A/x^2+B)
Sz = mat.╬╜ * (Sr + S╧Ж)

showElementResults(Sr, name="╧ГR")
showElementResults(S╧Ж, name="╧Г╧Ж")
showElementResults(Sz, name="╧Гz")
```

```julia
plotOnPath("bottom", sr)
plotOnPath("bottom", sz)
```
