```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
```

```julia
structured_box_mesh(y0=-0.5, lx=10, n=4)
```

```julia
gmsh.model.addPhysicalGroup(2, [2, 3, 4, 5, 6], -1, "free")
```

```julia
mat = Material("volume")
prob = Problem([mat], type=:VectorField);
```

```julia
K = symmetricGradientMatrix(prob) + gradDivMatrix(prob)
K[:, :]
```

```julia
ν = mat.ν
ex(x, y, z) = y / 100
eyz(x, y, z) = -ν * ex(x, y, z)
ε = TensorField(prob, "volume", [ex 0 0; 0 eyz 0; 0 0 eyz]);
```

```julia
g = -2 * (∇ ⋅ ε)        # vektor mező
f_vol = BoundaryCondition("volume", fx=g[1], fy=g[2], fz=g[3]);
```

```julia
ε_right = nodesToElements(elementsToNodes(ε), onPhysicalGroup="free")
n_right = normalVector(prob, "free")


t = 2 * (ε_right * n_right)
f_surf = BoundaryCondition("free", fx=t[1], fy=t[2], fz=t[3])
```

```julia
supp = BoundaryCondition("left", ux=0, uy=0, uz=0)
```

```julia
f = loadVector(prob, [f_vol, f_surf])
u = solveField(K, f, support=[supp])   # rigid mode kizárás
```

```julia
showDoFResults(u, name="u", visible=true, factor=10)
```

```julia
showDoFResults(ε, name="ε")
```

```julia
ε2 = (u ∘ ∇ + ∇ ∘ u) / 2
showDoFResults(ε2, name="ε2")
```

```julia
showDoFResults(ε - ε2, name="diff")
```
