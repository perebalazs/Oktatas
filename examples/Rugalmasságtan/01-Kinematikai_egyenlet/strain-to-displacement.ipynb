```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
```

```julia
structured_box_mesh(y0=-0.5, lx=10, n=4)
```

```julia
gmsh.model.addPhysicalGroup(2, [2, 3, 4, 5, 6], -1, "free")
```

```julia
mat = Material("volume")
prob = Problem([mat], type=:VectorField, field=:u, rhs_field=:f);
```

```julia
K = symmetricGradientMatrix(prob) + gradDivMatrix(prob)
K[:, :]
```

```julia
ОҪ = mat.ОҪ
ex(x, y, z) = y / 100
eyz(x, y, z) = -ОҪ * ex(x, y, z)
Оө = TensorField(prob, "volume", [ex 0 0; 0 eyz 0; 0 0 eyz]);
```

```julia
g = -2 * (вҲҮ вӢ… Оө)        # vektor mezЕ‘
f_vol = LoadCondition("volume", fx=g[1], fy=g[2], fz=g[3]);
```

```julia
Оө_right = nodesToElements(elementsToNodes(Оө), onPhysicalGroup="free")
n_right = normalVector(prob, "free")


t = 2 * (Оө_right * n_right)
f_surf = LoadCondition("free", fx=t[1], fy=t[2], fz=t[3])
```

```julia
supp = BoundaryCondition("left", ux=0, uy=0, uz=0)
```

```julia
f = loadVector(prob, [f_vol, f_surf])
u = solveField(K, f, support=[supp])   # rigid mode kizГЎrГЎs
```

```julia
showDoFResults(u, name="u", visible=true, factor=10)
```

```julia
showDoFResults(Оө, :ex, name="Оө")
```

```julia
Оө2 = (u вҲҳ вҲҮ + вҲҮ вҲҳ u) / 2
showDoFResults(Оө2, :ex, name="Оө2")
```

```julia
showDoFResults(Оө - Оө2, name="diff")
```
