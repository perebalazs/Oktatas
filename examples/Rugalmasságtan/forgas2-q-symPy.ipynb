```julia
using SymPy
```

```julia
function sx(s)
    x, y = symbols("x y")
    diff(diff(s, y), y)
end
function sy(s)
    x, y = symbols("x y")
    diff(diff(s, x), x)
end
function sxy(s)
    x, y = symbols("x y")
    -diff(diff(s, x), y)
end
function ΔΔ(f)
    x, y = symbols("x y")
    diff(diff(diff(diff(f, x), x), x), x) + 2 * diff(diff(diff(diff(f, x), x), y), y) + diff(diff(diff(diff(f, y), y), y), y)
end
function Δ(f)
    x, y = symbols("x y")
    diff(diff(f, x), x) + diff(diff(f, y), y)
end
```

```julia
function airy_ansatz(deg)
    x, y = symbols("x y")
    coeffs = Sym[]
    ϕ = 0

    k = 0
    for i in 0:deg
        for j in 0:(deg-i)
            C = symbols("C$k")
            push!(coeffs, C)
            ϕ += C * x^i * y^j
            k += 1
        end
    end

    return ϕ, coeffs
end
```

```julia
function coeff_equations(expr)
    x, y = symbols("x y")
    p = sympy.Poly(expr, x, y)
    return p.coeffs()
end
```

```julia
function coeff_equations(expr)
    x, y = symbols("x y")
    return collect(values(expr.as_poly(x, y).as_dict()))
end
```

```julia
ϕ, coeffs = airy_ansatz(6)
```

```julia
x, y = symbols("x y")
b, h = symbols("b h")
ρ, ω = symbols("ρ ω")
N, M, T, p, t, q = symbols("N M T p t q")
```

```julia
V = -ρ * ω^2 / 2 * (x^2 + y^2)
```

```julia
ΔΔϕ = ΔΔ(ϕ)
```

```julia
ΔV = Δ(V)
```

```julia
ν = symbols("ν")
```

```julia
eqs = coeff_equations(ΔΔϕ + (1 - ν) * ΔV)
```

```julia
σx = sx(ϕ) + V
```

```julia
σy = sy(ϕ) + V
```

```julia
τxy = sxy(ϕ)
```

```julia
s1 = σx(x => b / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
s1 = σx(x => -b / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
s1 = τxy(x => b / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
s1 = τxy(x => -b / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
s1 = σy(y => h / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
s1 = σy(y => -h / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
s1 = τxy(y => h / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
s1 = τxy(y => -h / 2)
eqs = vcat(eqs, coeff_equations(s1))
```

```julia
sol = solve(eqs, coeffs)
```

```julia
ϕ_sol = subs(ϕ, sol)
```

```julia
σx = sx(ϕ_sol) + V
factor(σx)
```

```julia
σy = sy(ϕ_sol) + V
```

```julia
τxy = sxy(ϕ_sol)
simplify(τxy)
```
