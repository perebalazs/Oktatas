```julia
using Plots
```

```julia
pk = 0.1
pb = 20

Rk = 40
Rb = 20

ρ = 30

δ = 0.002

E = 2e5
ν = 0.3

Ek = Eb = E
νk = νb = ν

par = Rb, Rk, ρ, pb, pk, Eb, Ek, νb, νk, δ
```

```julia
function pressure(par)
    Rb, Rk, ρ, pb, pk, Eb, Ek, νb, νk, δ = par

    A0k = Rk^2 + ρ^2
    Ak = Rk^2 - ρ^2

    A0b = ρ^2 + Rb^2
    Ab = ρ^2 - Rb^2

    B = A0k / Ak / Ek + A0b / Ab / Eb
    C = νk / Ek - νb / Eb
    D = Rk^2 * pk / Ak / Ek + Rb^2 * pb / Ab / Eb

    return (δ / ρ + 2 * D) / (B + C)
end
```

```julia
function σR(r, par)
    Rb, Rk, ρ, pb, pk, Eb, Ek, νb, νk, δ = par
    p = pressure(par)
    Ab = ρ^2 - Rb^2
    Ak = Rk^2 - ρ^2
    if r >= Rb && r <= ρ
        return (p - pb) * Rb^2 * ρ^2 / Ab / r^2 + (pb * Rb^2 - p * ρ^2) / Ab
    elseif r >= ρ && r <= Rk
        return (pk - p) * ρ^2 * Rk^2 / Ak / r^2 + (p * ρ^2 - pk * Rk^2) / Ak
    else
        return 0
        #error("σR: r is out of range")
    end
end

σR(r) = σR(r, par)
```

```julia
function σφ(r, par)
    Rb, Rk, ρ, pb, pk, Eb, Ek, νb, νk, δ = par
    p = pressure(par)
    Ab = ρ^2 - Rb^2
    Ak = Rk^2 - ρ^2
    if r >= Rb && r <= ρ
        return -(p - pb) * Rb^2 * ρ^2 / Ab / r^2 + (pb * Rb^2 - p * ρ^2) / Ab
    elseif r >= ρ && r <= Rk
        return -(pk - p) * ρ^2 * Rk^2 / Ak / r^2 + (p * ρ^2 - pk * Rk^2) / Ak
    else
        return 0
        #error("σφ: r is out of range")
    end
end

σφ(r) = σφ(r, par)
```

```julia
r = range(start=Rb, stop=Rk, length=1000)

plot(r, σR.(r))

plot!(r, σφ.(r))
```

```julia
ψ = Rb^2 ./ r .^ 2

plot(ψ, σR.(r))

plot!(ψ, σφ.(r))
```

```julia
plot(ψ, σφ.(r) .- σR.(r))
```

```julia
# Define the problem to solve
using Optimization, ForwardDiff, Zygote

function overlap(x, p)
    Rb, Rk, ρ, pb, pk, Eb, Ek, νb, νk, δ = p
    ρ = x[1]
    δ = x[2]
    par = Rb, Rk, ρ, pb, pk, Eb, Ek, νb, νk, δ
    srb = σR(Rb, par)
    sφb = σφ(Rb, par)
    srk = σR(ρ + 0.0001, par)
    sφk = σφ(ρ + 0.0001, par)
    mx = maximum(abs, [sφb - srb, sφk - srk])
    return mx
end





x0 = [30.0, 0.000001]
_p = par

#f = OptimizationFunction(overlap, Optimization.AutoForwardDiff())
#l1 = overlap(x0, _p)
#prob = OptimizationProblem(f, x0, _p)
```

```julia
using OptimizationBBO
prob = Optimization.OptimizationProblem(overlap, [30.0, 0.00001], _p, lb=[21.0, 0.0], ub=[39.0, 0.072])
sol = solve(prob, BBO_adaptive_de_rand_1_bin())
```

```julia
ρ = sol.u[1]
δ = sol.u[2]
par = Rb, Rk, ρ, pb, pk, Eb, Ek, νb, νk, δ
```
