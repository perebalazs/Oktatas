```julia
using Plots
```

```julia
pk = 0.1
pb = 20

Rk = 40
Rb = 20

ρ = 30

δ = 0.002

E = 2e5
ν = 0.3

Ek = Eb = E
νk = νb = ν
```

```julia
A0k = Rk^2 + ρ^2
Ak = Rk^2 - ρ^2

A0b = ρ^2 + Rb^2
Ab = ρ^2 - Rb^2

B = A0k / Ak / Ek + A0b / Ab / Eb
C = νk / Ek - νb / Eb
D = Rk^2 * pk / Ak / Ek + Rb^2 * pb / Ab / Eb

p = (δ / ρ + 2 * D) / (B + C)
```

```julia
σRb(r) = (p - pb) * Rb^2 * ρ^2 / Ab / r^2 + (pb * Rb^2 - p * ρ^2) / Ab
σRk(r) = (pk - p) * ρ^2 * Rk^2 / Ak / r^2 + (p * ρ^2 - pk * Rk^2) / Ak

σφb(r) = -(p - pb) * Rb^2 * ρ^2 / Ab / r^2 + (pb * Rb^2 - p * ρ^2) / Ab
σφk(r) = -(pk - p) * ρ^2 * Rk^2 / Ak / r^2 + (p * ρ^2 - pk * Rk^2) / Ak
```

```julia
rb = range(start=Rb, stop=ρ, length=10)
rk = range(start=ρ, stop=Rk, length=10)

plot(rb, σRb.(rb))
plot!(rk, σRk.(rk))

plot!(rb, σφb.(rb))
plot!(rk, σφk.(rk))
```

```julia
ψb = Rb^2 ./ rb .^ 2
ψk = Rb^2 ./ rk .^ 2

plot(ψb, σRb.(rb))
plot!(ψk, σRk.(rk))

plot!(ψb, σφb.(rb))
plot!(ψk, σφk.(rk))
```
