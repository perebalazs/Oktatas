```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()
```

```julia
mat = Material("body")

PF = Problem([mat], type=:ScalarField, dim=2, field=:F, rhs_field=:F)
Pp = Problem([mat], type=:ScalarField, dim=2, field=:p, rhs_field=:p)
```

```julia
KF = вҲ«(Grad(PF) вӢ… Grad(PF))
KFp = вҲ«(PF вӢ… Pp)
KF2 = вҲ«(Grad(PF) вӢ… ([0, 1] * [0, 1]') вӢ… Grad(PF), О“="right")
KF3 = вҲ«(Grad(PF) вӢ… ([1, 0] * [0, 1]') вӢ… Grad(PF), О“="right")

Kpp = вҲ«(Grad(Pp) вӢ… Grad(Pp))

K = SystemMatrix([
    KF+KF2+KF3 KFp
    0KFp' Kpp])
K[:, :]
```

```julia
qF = -вҲ«(PF вӢ… 1, О“="right")
qp = -вҲ«(Pp вӢ… 0)
```

```julia
bcF_1 = BoundaryCondition("leftbottom", problem=PF, F=0)
bcF_2 = BoundaryCondition("rightbottom", problem=PF, F=0)
bcF_3 = BoundaryCondition("lefttop", problem=PF, F=0)

bcp_1 = BoundaryCondition("leftbottom", problem=Pp, p=0)

bc = [
    bcF_1, bcF_2, bcF_3,
    bcp_1
]
```

```julia
q = SystemVector([qF, qp])   # vagy nullvektor a megfelelЕ‘ mГіdon

F, p = solveField(K, q, support=bc)

showDoFResults(F, name="F")
showDoFResults(p, name="p")
```
