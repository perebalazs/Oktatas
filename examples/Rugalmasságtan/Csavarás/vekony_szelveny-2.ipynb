```julia
using LowLevelFEM
```

```julia
#structured_rect_mesh(lx=10, n=10, order=1)
openGeometry("vekonyszelveny1.geo")

openPreProcessor()

mat = Material("body")
Pφ = Problem([mat], type=:ScalarField, dim=2, field=:φ, rhs_field=:p)
```

```julia
μ = mat.μ

Mc = 1
```

```julia
K = ∫(Grad(Pφ) ⋅ Grad(Pφ))

f = ∫(Pφ ⋅ (-2))
```

```julia
bc = BoundaryCondition("perem", φ=0)
```

```julia
ψ = solveField(K, f, support=[bc])
```

```julia
showDoFResults(ψ, name="ψ", visible=true)
```

```julia
Ic = 2 * ∫(Pφ, "body", ψ)
```

```julia
β = Mc / (μ * Ic)
```

```julia
φ = μ * β * ψ

Φ = VectorField(Pφ, "body", [0, 0, φ])
```

```julia
τ = rot(Φ)
```

```julia
showDoFResults(τ, name="τ")
```
