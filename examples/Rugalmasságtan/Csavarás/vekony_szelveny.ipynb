```julia
using LowLevelFEM
```

```julia
structured_rect_mesh(lx=10, n=10, order=1)

openPreProcessor()

mat = Material("body")
Pφ = Problem([mat], type=:ScalarField, dim=2, field=:φ, rhs_field=:p)
```

```julia
μ = mat.μ

Mc = 1
```

```julia
K = ∫(Grad(Pφ) ⋅ Grad(Pφ))

f = ∫(Pφ ⋅ (-2))
```

```julia
bcL = BoundaryCondition("left", φ=0)
bcR = BoundaryCondition("right", φ=0)
bcB = BoundaryCondition("bottom", φ=0)
bcT = BoundaryCondition("top", φ=0)
```

```julia
ψ = solveField(K, f, support=[bcL, bcR, bcB, bcT])
ψ2 = solveField(K, f, support=[bcB, bcT])
```

```julia
showDoFResults(ψ, name="ψ", visible=true)
showDoFResults(ψ2, name="ψ2")
```

```julia
Ic = 2 * ∫(Pφ, "body", ψ)
```

```julia
Ic2 = 2 * ∫(Pφ, "body", ψ2)
```

```julia
β = Mc / (μ * Ic)
```

```julia
β2 = Mc / (μ * Ic2)
```

```julia
φ = μ * β * ψ

Φ = VectorField(Pφ, "body", [0, 0, φ])
```

```julia
φ2 = μ * β * ψ2

Φ2 = VectorField(Pφ, "body", [0, 0, φ2])
```

```julia
τ = rot(Φ)
```

```julia
τ2 = rot(Φ2)
```

```julia
t1 = showDoFResults(τ, :x, name="τ")
showDoFResults(τ2, name="τ2")
```

```julia
plotOnPath("path", t1)
```
