```julia
using LowLevelFEM, LinearAlgebra
```

```julia
structured_box_mesh(lx=10, order=1)

mat = Material("body")
Pu = Problem([mat])
```

```julia
Mc = 1
```

```julia
bcL = BoundaryCondition("left", ux=0, uy=0, uz=0)

c = 0.01 / 11.18
fy(x, y, z) = -(z - 0.5) * c
fz(x, y, z) = (y - 0.5) * c
bcR = BoundaryCondition("right", uy=fy, uz=fz)
```

```julia
u = solveDisplacement(Pu, support=[bcL, bcR], iterative=true)
showDoFResults(u, name="u")
```

```julia
S = solveStress(u)
n = VectorField(Pu, "body", [1, 0, 0])
τ = S * n
c1 = ScalarField(Pu, "body", 1)
c2 = ScalarField(Pu, "left", 1)
c2 = nodesToElements(elementsToNodes(c2))
c = c1 - c2
showDoFResults(c)
showDoFResults(S * n * c, name="σ")
```

```julia
S = nodesToElements(elementsToNodes(S), onPhysicalGroup="right")
r = nodesToElements(nodePositionVector(Pu), onPhysicalGroup="right")
n = normalVector(Pu, "right")

rxτ = r × (S * n)
Mc = ∫(Pu, "right", rxτ[1])
```
