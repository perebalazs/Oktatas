```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()
#openPreProcessor()

mat = Material("surface", E=2e5, ОҪ=0.3)
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f);
```

```julia
E = mat.E
ОҪ = mat.ОҪ
```

```julia
D = E / (1 - ОҪ^2) * [1 ОҪ 0; ОҪ 1 0; 0 0 (1-ОҪ)/2]
```

```julia
D = E / (1 + ОҪ) / (1 - 2ОҪ) * [1-ОҪ ОҪ 0; ОҪ 1-ОҪ 0; 0 0 (1-2ОҪ)/2]
```

```julia
K = вҲ«(SymGrad(Pu) вӢ… D вӢ… SymGrad(Pu), О©="surface")
K[:, :]
```

```julia
ff = ScalarField(Pu, "surface", 1)
fx = [ff, 1]
```

```julia
f = вҲ«(Pu вӢ… fx, О“="right")
```

```julia
suppL = BoundaryCondition("left", ux=0)
suppB = BoundaryCondition("bottom", uy=0);
```

```julia
u = solveField(K, f, support=[suppL, suppB])
```

```julia
showDoFResults(u, name="u")
```

```julia
Оөx = вҲӮx(u[1])
Оөy = вҲӮy(u[2])
Оіxy = вҲӮy(u[1]) + вҲӮx(u[2])

Оө = [Оөx, Оөy, Оіxy];
```

```julia
Пғ = D * Оө;
```

```julia
showElementResults(Пғ[1], name="Пғx")
showElementResults(Пғ[1], name="Пғx", smooth=true)
showElementResults(Пғ[2], name="Пғy")
showElementResults(Пғ[3], name="Оіxy")
```
