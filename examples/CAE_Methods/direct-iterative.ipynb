```julia
using LowLevelFEM
using LinearAlgebra
```

```julia
structured_box_mesh(n=30, order=1)

mat = Material("body")
prob = Problem([mat])
```

```julia
K = stiffnessMatrix(prob)
K[:, :]
```

```julia
supp = displacementConstraint("left", ux=0, uy=0, uz=0)

ld = load("right", fy=1)
```

```julia
println("DOF = ", prob.non * prob.pdim)
```

```julia
@time u1 = solveDisplacement(prob, support=[supp], load=[ld])
```

```julia
@time u2 = solveDisplacement(prob, support=[supp], load=[ld], iterative=true, reltol=1e-2)
```

```julia
err = norm(DoFs(u1 - u2)) / norm(DoFs(u1))
println("Relative error: ", err)
```
