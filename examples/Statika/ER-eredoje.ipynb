```julia
using LowLevelFEM, LinearAlgebra

gmsh.initialize()
gmsh.open("ER1.geo")
```

```julia
mat = material("frame")
prob = Problem([mat], type=:Truss)
```

```julia
FA = load("A", fx=1, fy=3)
FB = load("B", fy=-1)
FD = load("D", fz=1)
```

```julia
F_ER1 = loadVector(prob, [FA, FB, FD]);
```

```julia
showDoFResults(F_ER1, name="ER")
```

```julia
F_er = resultant(F_ER1, "frame")
```

```julia
r = nodePositionVector(prob)

M_O = r × F_ER1;
```

```julia
M_er = resultant(elementsToNodes(M_O), "frame")
```

```julia
F_er = vectorField(prob, [field("O", fx=F_er[1], fy=F_er[2], fz=F_er[3])]);
```

```julia
M_er = vectorField(prob, [field("O", fx=M_er[1], fy=M_er[2], fz=M_er[3])]);
```

```julia
showDoFResults(F_er, name="F eredő")
showDoFResults(M_er, name="M eredő")
```
