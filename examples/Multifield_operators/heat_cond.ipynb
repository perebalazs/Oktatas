```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()

mat = Material("surface")
PT = Problem([mat], type=:ScalarField, dim=2, field=:T, rhs_field=:q)
```

```julia
k = mat.k
```

```julia
K = вҲ«(Grad(PT) вӢ… Grad(PT) * k)
K[:, :]
```

```julia
q = вҲ«(PT вӢ… 100)
```

```julia
supp = BoundaryCondition("left", T=0)
```

```julia
T = solveField(K, q, support=[supp])
```

```julia
showDoFResults(T)
```

```julia
Kh = вҲ«(PT вӢ… PT * h, О“="right")
Kh[:, :]
```

```julia
qh = вҲ«(PT вӢ… (h * TвҲһ), О“="right")
```

```julia
T2 = solveField(K + Kh, qh, support=[supp])
```

```julia
showDoFResults(T2)
```
