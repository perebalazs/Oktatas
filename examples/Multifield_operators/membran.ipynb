```julia
using LowLevelFEM, LinearAlgebra
```

```julia
#structured_rect_mesh()
openGeometry("shell.geo")
openPreProcessor()

mat = Material("body")
Pu = Problem([mat], type=:VectorField, dim=3, field=:u, rhs_field=:f)
```

```julia
E = mat.E
ν = mat.ν

Cs = E / (1 - ν^2) * [1 ν 0; ν 1 0; 0 0 (1-ν)/2]
K = ∫(SurfaceSymGrad(Pu) ⋅ Cs ⋅ SurfaceSymGrad(Pu); Γ="body")
K[:, :]
```

```julia
rank(K.A)
```

```julia
vals = eigvals(Matrix(K.A))
```
