```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()

mat = Material("body")
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f)
PT = Problem([mat], type=:ScalarField, dim=2, field=:T, rhs_field=:q)
```

```julia
E = mat.E
ν = mat.ν
k = mat.k / 10
α = mat.α
κ = mat.κ
ρ = mat.ρ
c = mat.c

#T0 = 293

dt = 0.001
ϑ = 0.5

D = E / (1 - ν^2) * [1 ν 0; ν 1 0; 0 0 (1-ν)/2]
```

```julia
Dv = D / 100
```

```julia
Ku = ∫(SymGrad(Pu) ⋅ D ⋅ SymGrad(Pu))
```

```julia
Kv = ∫(SymGrad(Pu) ⋅ Dv ⋅ SymGrad(Pu))
```

```julia
f = ∫(Pu ⋅ [0, -1], Γ="right")
```

```julia
KT = ∫(Grad(PT) ⋅ k ⋅ Grad(PT))
```

```julia
CT = ∫(PT ⋅ (c * ρ) ⋅ PT)
```

```julia
q = ∫(PT ⋅ 0)
```

```julia
supp_u = BoundaryCondition("left", problem=Pu, ux=0, uy=0)
```

```julia
u = solveField(Ku, 200f, support=[supp_u])
```

```julia
showDoFResults(u, name="u")
```

```julia
u0 = 0f
u1 = 0f
T0 = 0q

εx = ∂x(u0[1])
εy = ∂y(u0[2])
γxy = ∂y(u0[1]) + ∂x(u0[2])

ε0 = [εx, εy, γxy]
```

```julia
diss = nothing
T1 = nothing

uu = VectorField[]
dd = ScalarField[]
TT = ScalarField[]

for i in 1:200
    u1 = solveField(Ku + Kv / dt, Kv * u0 / dt + i * f, support=[supp_u])
    push!(uu, copy(u1))

    εx = ∂x(u1[1])
    εy = ∂y(u1[2])
    γxy = ∂y(u1[1]) + ∂x(u1[2])
    ε1 = [εx, εy, γxy]

    dε = (ε1 - ε0) / dt
    diss = dε' * Dv * dε
    push!(dd, elementsToNodes(copy(diss)))

    q_diss = ∫(PT ⋅ diss)

    #T1 = (CT + dt * ϑ * KT) \ ((CT - dt * (1 - ϑ) * KT) * T0 + dt * q_diss)
    T1 = solveField(CT + dt * ϑ * KT, (CT - dt * (1 - ϑ) * KT) * T0 + dt * q_diss)
    push!(TT, copy(T1))

    u0 = copy(u1)
    ε0 = copy(ε1)
    T0 = copy(T1)
end
```

```julia
u2 = 0f
uu2 = VectorField[]

for i in 1:100
    u2 = solveField(Ku + Kv / dt, Kv * u0 / dt + 200 * f, support=[supp_u])
    u0 = copy(u2)
    push!(uu2, copy(u2))
end
uu2 = mergeFields(uu2)
```

```julia
uu = mergeFields(uu)
dd = mergeFields(dd)
TT = mergeFields(TT)
```

```julia
showDoFResults(uu, name="uu")
showDoFResults(dd, name="diss")
showDoFResults(TT, name="TT")
showDoFResults(εx, name="εx")
```
