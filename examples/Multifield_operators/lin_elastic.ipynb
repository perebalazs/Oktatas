```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()

mat = Material("surface")
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f)
```

```julia
E = mat.E
ОҪ = mat.ОҪ

D = E / (1 + ОҪ^2) * [1 ОҪ 0; ОҪ 1 0; 0 0 (1+ОҪ)/2]
```

```julia
K = вҲ«(SymGrad(Pu) вӢ… D вӢ… SymGrad(Pu))
K[:, :]
```

```julia
f = вҲ«(Pu вӢ… [1, 0], О“="right")
```

```julia
supp1 = BoundaryCondition("left", ux=0)
supp2 = BoundaryCondition("bottom", uy=0)
```

```julia
u = solveField(K, f, support=[supp1, supp2])
```

```julia
showDoFResults(u)
```
