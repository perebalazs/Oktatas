```julia
using LowLevelFEM
```

```julia
line_mesh()

mat = Material("body")
Pu = Problem([mat], type=:ScalarField, dim=3, field=:u, rhs_field=:N)
Pv = Problem([mat], type=:ScalarField, dim=2, field=:v, rhs_field=:T)
Pφ = Problem([mat], type=:ScalarField, dim=2, field=:φ, rhs_field=:M)
```

```julia
E = mat.E
G = mat.μ
I = 1
κ = 2 / 3
A = 1
```

```julia
Kuu = ∫(AxialGrad(Pu) ⋅ (E * A) ⋅ AxialGrad(Pu); Γ="body")

Kvv = ∫(AxialGrad(Pv) ⋅ (κ * G * A) ⋅ AxialGrad(Pv); Γ="body")

Kvφ = ∫(AxialGrad(Pv) ⋅ (-κ * G * A) ⋅ Id(Pφ); Γ="body")

Kφv = ∫(Id(Pφ) ⋅ (-κ * G * A) ⋅ AxialGrad(Pv); Γ="body")

Kφφ = ∫(Id(Pφ) ⋅ (κ * G * A) ⋅ Id(Pφ); Γ="body") +
      ∫(AxialGrad(Pφ) ⋅ (E * I) ⋅ AxialGrad(Pφ); Γ="body")
```

```julia
Kuu = ∫(AxialGrad(Pu) ⋅ Duu ⋅ AxialGrad(Pu); Γ="body")

Kuφ = ∫(AxialGrad(Pu) ⋅ Duφ ⋅ Id(Pφ); Γ="body")

Kφu = ∫(Id(Pφ) ⋅ Dφu ⋅ AxialGrad(Pu); Γ="body")

Kφφ = ∫(Id(Pφ) ⋅ Dφφ_id ⋅ Id(Pφ); Γ="body") +
      ∫(AxialGrad(Pφ) ⋅ Dφφ_grad ⋅ AxialGrad(Pφ); Γ="body")

K = SystemMatrix([
    Kuu Kuφ
    Kφu Kφφ
])
```
