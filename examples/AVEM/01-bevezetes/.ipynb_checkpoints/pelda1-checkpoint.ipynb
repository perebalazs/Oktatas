```julia
using LowLevelFEM
gmsh.initialize()
```

```julia
gmsh.open("cube.geo")
```

```julia
mat = material("cube", E=2e5, Î½=0.3)
prob = Problem([mat])
```

```julia
load1 = load("load", fy=-1)
supp = displacementConstraint("supp", ux=0, uy=0, uz=0)
```

```julia
u = solveDisplacement(prob, [load1], [supp])
```

```julia
showDoFResults(u, :uvec)
```

```julia
S = solveStress(u)
```

```julia
sx = showStressResults(S, :sx)
```

```julia
plotOnPath("path", sx)
```

```julia
openPostProcessor()
```

```julia
gmsh.finalize()
```
