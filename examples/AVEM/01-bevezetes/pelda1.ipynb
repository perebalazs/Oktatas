```julia
using LowLevelFEM
gmsh.initialize()
```

```julia
gmsh.open("cube.geo")
```

```julia
mat = material("cube", E=2e5, ν=0.3)
prob = Problem([mat])
```

```julia
load1 = load("load", fy=-1)
supp = displacementConstraint("supp", ux=0, uy=0, uz=0)
```

```julia
stiffnessMatrix(prob)
```

```julia
u = solveDisplacement(prob, [load1], [supp])
```

```julia
showDoFResults(u, :uvec)
```

```julia
@time A1 = LowLevelFEM.solveStrainOld(u)
```

```julia
@time A2 = solveStrain(u)
```

```julia
S = solveStress(u)
```

```julia
ex1 = showStrainResults(A1, :ex)
ex2 = showStrainResults(A2, :ex)
```

```julia
sx = showStressResults(S, :sx)
```

```julia
plotOnPath("path", sx)
```

```julia
openPostProcessor()
```

```julia
gmsh.finalize()
```
