```julia
using LowLevelFEM
gmsh.initialize()
```

```julia
gmsh.open("rod.geo")
```

```julia
mat = material("rod", E=2e5)
prob = Problem([mat], type=:Truss)
```

```julia
load1 = load("load", fx=1)
supp0 = displacementConstraint("rod", uy=0, uz=0)
supp = displacementConstraint("supp", ux=0)
```

```julia
u = solveDisplacement(prob, [load1], [supp0, supp])
```

```julia
showDoFResults(u, :uvec)
```

```julia
openPostProcessor()
```

```julia
gmsh.finalize()
```
