```julia
using LowLevelFEM
gmsh.initialize()

gmsh.open("truss2.geo")
```

```julia
mat = material("rods", E=2e5, A=100)
prob = Problem([mat], type=:Truss)
```

```julia
supp1 = displacementConstraint("left", ux=0, uy=0)
supp2 = displacementConstraint("right", uy=0)
load1 = load("force1", fy=-100)

supp0 = displacementConstraint("rods", uz=0);
```

```julia
q = solveDisplacement(prob, [load1], [supp0, supp1, supp2])
```

```julia
N = solveAxialForce(q)
N.A
```

```julia
N.numElem
```

```julia
r = nodePositionVector(prob)
r = nodesToElements(r)
```

```julia
r.A[3]
```

```julia
ux = showDoFResults(q)
N0 = showElementResults(N)
```

```julia
K = stiffnessMatrix(prob)
```

```julia
f = K * q
```
