```julia
using LowLevelFEM, InteractiveUtils

@trace_compile begin
    gmsh.initialize()
    gmsh.open("cube.geo")

    mat = material("body")
    prob = Problem([mat])

    K = stiffnessMatrix(prob)

    load1 = load("front", fz=1)
    supp = displacementConstraint("rear", ux=0, uy=0, uz=0)

    q = solveDisplacement(prob, [load1], [supp])
    gmsh.finalize()
end
```
