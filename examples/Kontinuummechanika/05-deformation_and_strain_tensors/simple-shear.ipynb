```julia
using LowLevelFEM
gmsh.initialize()
```

```julia
gmsh.open("cube.geo")
```

```julia
mat = material("body")
prob = Problem([mat])
```

```julia
γ = 0.5
x(X, Y, Z) = X + γ * Y
y(X, Y, Z) = Y
z(X, Y, Z) = Z
```

```julia
r = vectorField(prob, "body", [x, y, z]);
```

```julia
r0 = showDeformationResults(r, visible=true, name="F")
```

```julia
F = r ∘ ∇;
I = unitTensor(F);
```

```julia
R = nodePositionVector(prob);
```

```julia
U = √(F' * F);
showDeformationResults(U * R, name="U")
```

```julia
C = F' * F
showDeformationResults(C * R, name="C")
```

```julia
E = (F' * F - I) / 2
showDoFResults(E * R, name="E")
```

```julia
H = log(U)
showDoFResults(H * R, name="H")
```

```julia
b = F * F'
showDeformationResults(b * R, name="b")
```

```julia
showDeformationResults(inv(b) * r, name="inv(b)")
```

```julia
e = (I - inv(b)) / 2
showDoFResults(e * r, name="e")
```

```julia
v = √(F * F')
showDeformationResults(v * R, name="v")
```

```julia
h = log(v)
showDoFResults(h * R, name="h")
```

```julia
D = (F - I)
showDoFResults(D * R, name="D")
```

```julia
A = (D + D') / 2
showDoFResults(A * R, name="A")
```

```julia
Φ = (D - D') / 2
showDoFResults(Φ * R, name="Φ")
```

```julia
Rot = F * inv(U)
showDeformationResults(Rot * R, name="R")
```

```julia
J = det(F)
Uvol = (∛J) * I
showDeformationResults(Uvol * R, name="Uvol")
```

```julia
Udev = U / ∛J
showDeformationResults(Udev * R, name="Udev")
```
