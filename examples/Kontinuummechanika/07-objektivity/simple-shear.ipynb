```julia
using LowLevelFEM, LinearAlgebra
```

```julia
gmsh.initialize()
gmsh.open("cube.geo")
```

```julia
mat = material("body")
prob = Problem([mat])
```

```julia
γ = 0.5
x(X, Y, Z) = X + γ * Y
y(X, Y, Z) = Y
z(X, Y, Z) = Z
```

```julia
R = nodePositionVector(prob)
r = vectorField(prob, "body", [x, y, z]);
```

```julia
r0 = showDeformationResults(r, name="F")
```

```julia
F = r ∘ ∇;
I = unitTensor(F);
```

```julia
rnd1 = rand(3)
v1 = VectorField(prob, "body", rnd1 / norm(rnd1))
rnd2 = rand(3)
v2 = VectorField(prob, "body", rnd2 / norm(rnd2))
v2 = v1 × v2 / norm(v1 × v2)
v3 = v1 × v2

ex = VectorField(prob, "body", [1, 0, 0])
ey = VectorField(prob, "body", [0, 1, 0])
ez = VectorField(prob, "body", [0, 0, 1])

Q = v1 ∘ ex + v2 ∘ ey + v3 ∘ ez;
```

```julia
showDoFResults(v1, name="n1")
showDoFResults(v2, name="n2")
showDoFResults(v3, name="n3")
```

```julia
F⁺ = Q * F

showDeformationResults(F⁺ * R, name="F⁺")
```

```julia
C = F' * F
C⁺ = F⁺' * F⁺

showDeformationResults(C * R, name="C")
showDeformationResults(C⁺ * R, name="C⁺")
```

```julia
b = F * F'
b⁺ = F⁺ * F⁺'

showDeformationResults(b * R, name="b")
showDeformationResults(b⁺ * R, name="b⁺")
```
