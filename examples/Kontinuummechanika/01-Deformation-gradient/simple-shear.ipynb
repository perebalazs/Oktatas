```julia
using LowLevelFEM
gmsh.initialize()
```

```julia
gmsh.open("cube.geo")
```

```julia
mat = material("body")
problem = Problem([mat])
```

```julia
γ = 0.5
x(X, Y, Z) = X + γ * Y
y(X, Y, Z) = Y
z(X, Y, Z) = Z
```

```julia
deformation = field("body", fx=x, fy=y, fz=z)
r = vectorField(problem, [deformation])
```

```julia
showDeformationResults(r, :vector, visible=true)
```

```julia
openPostProcessor()
```

```julia
#F = ∇(r)
F = r ∘ ∇
```

```julia
atP = "P"
FatP = probe(F, atP)
```

```julia
R = nodePositionVector(problem)

R = nodesToElements(R)

showDeformationResults(F * R, :vector)
```

```julia
r = nodesToElements(r)
showDeformationResults(inv(F) * r, :vector)
```
