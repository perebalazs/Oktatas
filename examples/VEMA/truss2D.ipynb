```julia
using LowLevelFEM
```

```julia
openGeometry("truss2D.geo")

#openPreProcessor(openGL=true)
```

```julia
mat = Material("rod", E=1, A=1)
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f)
```

```julia
A = mat.A
E = mat.E
```

```julia
t = tangentVector(Pu, "rod")
showElementResults(t)
t1 = [t[1]; t[2]; 0; 0;;]
```

```julia
t1 * t1'
```

```julia
K = ∫(Grad(Pu) ⋅ (t1 * t1') ⋅ Grad(Pu))
K[:, :]
```

```julia
#f0 = ∫(Pu ⋅ [1,1])
```

```julia
fR = ∫(Pu ⋅ [0, -1], Γ="3")
```

```julia
supp1 = BoundaryCondition("1", ux=0, uy=0)
supp2 = BoundaryCondition("2", ux=0, uy=0)
```

```julia
u = solveField(K, fR, support=[supp1, supp2])
u.a
```
