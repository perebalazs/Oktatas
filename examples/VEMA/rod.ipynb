```julia
using LowLevelFEM
```

```julia
openGeometry("rod.geo")

#openPreProcessor(openGL=true)
```

```julia
mat = Material("rod", E=1, A=1)
Pu = Problem([mat], type=:ScalarField, dim=1, field=:u, rhs_field=:f)
```

```julia
A = mat.A
E = mat.E
```

```julia
K = ∫(AxialGrad(Pu) ⋅ (1) ⋅ AxialGrad(Pu))
K[:, :]
```

```julia
f0 = ∫(Pu ⋅ 1)
```

```julia
fR = ∫(Pu ⋅ 10, Γ="right")
```

```julia
f = f0 + fR
```

```julia
supp = BoundaryCondition("left", u=0)
```

```julia
u = solveField(K, f, support=[supp])
```

```julia
N = A * E * ∂x(u)
```

```julia
u0 = showDoFResults(u)
N0 = showElementResults(N)
```

```julia
plotOnPath("rod", u0)
plotOnPath("rod", N0)
```
