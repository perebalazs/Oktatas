```julia
using LowLevelFEM
```

```julia
line_mesh(lx=100, n=100)

mat = Material("body", E=2e5, ПҒ=7.85e-9, A=1)
Pu = Problem([mat], type=:ScalarField, dim=1, field=:u)
```

```julia
E = mat.E
ПҒ = mat.ПҒ

K = вҲ«(Grad(Pu) вӢ… E вӢ… Grad(Pu))
K[:, :]
```

```julia
M = вҲ«(Pu вӢ… ПҒ вӢ… Pu)
M = consistentToLumped(M)
M[:, :]
```

```julia
f = вҲ«(Pu вӢ… 0)
```

```julia
bc = BoundaryCondition("right", u=0)

u0 = scalarField(Pu, "body", 0)
v0 = scalarField(Pu, "body", 1)
v1 = scalarField(Pu, "right", 1)
v0 -= v1
```

```julia
О»max = largestEigenValue(K, M, support=[bc])

dt = 2 / вҲҡО»max # Tmax / ПҖ
```

```julia
u, v = CDM(K, M, f, u0, v0, 300, dt, support=[bc])
```

```julia
u1 = showDoFResults(u, name="u")
v1 = showDoFResults(v, name="v")

plotOnPath("body", u1)
plotOnPath("body", v1)
```
