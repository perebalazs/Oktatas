```julia
using LowLevelFEM
gmsh.initialize()
```

```julia
gmsh.open("cylinder.geo")
```

```julia
mat = material("body")
prob = Problem([mat])
```

```julia
ux(x, y, z) = -y * z / 10000
uy(x, y, z) = x * z / 10000

fl = field("body", fx=ux, fy=uy)
u = vectorField(prob, [fl])
```

```julia
showDoFResults(u, :uvec, visible=true)
```

```julia
S = solveStress(u)
```

```julia
showStressResults(S, :s)
```

```julia
using LinearAlgebra
λ, Q = eigen(S)
```

```julia
λ1, λ2, λ3 = getEigenValues(λ)
```

```julia
v1, v2, v3 = getEigenVectors(Q)
```

```julia
showDoFResults(λ1, :scalar, name="λ1")
showDoFResults(λ2, :scalar, name="λ2")
showDoFResults(λ3, :scalar, name="λ3")
```

```julia
showDoFResults(v1, :vector, name="v1")
showDoFResults(v2, :vector, name="v2")
showDoFResults(v3, :vector, name="v3")
```
