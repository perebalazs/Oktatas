```julia
using LowLevelFEM
gmsh.initialize()
```

```julia
gmsh.open("designe-truss.geo")
```

```julia
mat = material("plate")
prob = Problem([mat], type=:PlaneStress)
```

```julia
load1 = load("load", fy=-1)
supp1 = displacementConstraint("left", ux=0, uy=0)
supp2 = displacementConstraint("right", uy=0)
```

```julia
u = solveDisplacement(prob, [load1], [supp1, supp2])
```

```julia
showDoFResults(u, :uvec, visible=true)
```

```julia
S = solveStress(u)
```

```julia
s = showStressResults(S, :s)
```

```julia
using LinearAlgebra
λ, Q = eigen(S)
```

```julia
λ1, λ2, λ3 = getEigenValues(λ)
```

```julia
v1, v2, v3 = getEigenVectors(Q)
```

```julia
isElementwise(v1)
```

```julia
v1 = elementsToNodes(v1)
v2 = elementsToNodes(v2)
v3 = elementsToNodes(v3)
```

```julia
isNodal(v1)
```

```julia
v1 = projectTo2D(v1)
v2 = projectTo2D(v2)
v3 = projectTo2D(v3)
```

```julia
showDoFResults(λ1, :scalar, name="λ1")
showDoFResults(λ2, :scalar, name="λ2")
showDoFResults(λ3, :scalar, name="λ3")
```

```julia
showDoFResults(v1, :vector)
showDoFResults(v2, :vector)
showDoFResults(v3, :vector)
```
