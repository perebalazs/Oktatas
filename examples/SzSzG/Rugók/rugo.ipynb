```julia
using LowLevelFEM, LinearAlgebra
```

```julia
gmsh.initialize()
gmsh.open("rud-negyzet.geo")
```

```julia
mat = material("rud")
prob = Problem([mat])

E = prob.material[1].E
ν = prob.material[1].ν

I = TensorField(prob, "rud", [1 0 0; 0 1 0; 0 0 1]);
```

```julia
Rp02 = 100

F1 = TensorField(prob, "rud", [0 0 0; 0 0 0; 0 0 Rp02]);
```

```julia
A1 = (1 + ν) / E * (F1 - ν / (1 + ν) * trace(F1) * I);
```

```julia
u1 = F1 ⋅ A1 / 2;
```

```julia
U1 = integrate(prob, "rud", u1)
```

```julia
f(x, y, z) = 1
V1 = integrate(prob, "rud", f)
```

```julia
E1 = U1 / V1
```

```julia
Rp022a(x, y, z) = y / 5 * Rp02
F2a = TensorField(prob, "rud", [0 0 0; 0 0 0; 0 0 Rp022a])
A2a = (1 + ν) / E * (F2a - ν / (1 + ν) * trace(F2a) * I)
u2a = F2a ⋅ A2a / 2
U2a = integrate(prob, "rud", u2a)
V2a = integrate(prob, "rud", f)
E2a = U2a / V2a
```

```julia
Rp022b(x, y, z) = y / 5 * Rp02 * (100 - z) / 100
F2b = TensorField(prob, "rud", [0 0 0; 0 0 0; 0 0 Rp022b])
A2b = (1 + ν) / E * (F2b - ν / (1 + ν) * trace(F2b) * I)
u2b = F2b ⋅ A2b / 2
U2b = integrate(prob, "rud", u2b)
V2b = integrate(prob, "rud", f)
E2b = U2b / V2b
```

```julia
E2a / E1
```

```julia
E2b / E1
```

```julia
gmsh.open("rud-kor.geo")

prob2 = Problem([mat])

I = TensorField(prob2, "rud", [1 0 0; 0 1 0; 0 0 1]);
```

```julia
Rp022a2(x, y, z) = y / 5 * Rp02
F2a2 = TensorField(prob2, "rud", [0 0 0; 0 0 0; 0 0 Rp022a2])
A2a2 = (1 + ν) / E * (F2a2 - ν / (1 + ν) * trace(F2a2) * I)
u2a2 = F2a2 ⋅ A2a2 / 2
U2a2 = integrate(prob2, "rud", u2a2)
V2a2 = integrate(prob2, "rud", f)
E2a2 = U2a2 / V2a2
```

```julia
Rp022b2(x, y, z) = y / 5 * Rp02 * (100 - z) / 100
F2b2 = TensorField(prob2, "rud", [0 0 0; 0 0 0; 0 0 Rp022b2])
A2b2 = (1 + ν) / E * (F2b2 - ν / (1 + ν) * trace(F2b2) * I)
u2b2 = F2b2 ⋅ A2b2 / 2
U2b2 = integrate(prob2, "rud", u2b2)
V2b2 = integrate(prob2, "rud", f)
E2b2 = U2b2 / V2b2
```

```julia
Rp023x(x, y, z) = -y / 5 * Rp02 / √3
Rp023y(x, y, z) = x / 5 * Rp02 / √3
F3 = TensorField(prob2, "rud", [0 0 Rp023x; 0 0 Rp023y; Rp023x Rp023y 0])
A3 = (1 + ν) / E * (F3 - ν / (1 + ν) * trace(F3) * I)
u3 = F3 ⋅ A3 / 2
U3 = integrate(prob2, "rud", u3)
V3 = integrate(prob2, "rud", f)
E3 = U3 / V3
```

```julia
E2a2 / E1
```

```julia
E2b2 / E1
```

```julia
E3 / E1
```
