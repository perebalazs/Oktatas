```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("lancszem.geo")
```

```julia
mat = material("disk")
prob = Problem([mat], type=:PlaneStress)
```

```julia
f(x, y, z) = y
f1(x, y, z) = 1
f2(x, y, z) = 1 / y
```

```julia
Sx = integrate(prob, "disk", f)
```

```julia
A = integrate(prob, "disk", f1)
```

```julia
Sx / A
```

```julia
I = integrate(prob, "disk", f2, order=1)
```

```julia
A / I
```
