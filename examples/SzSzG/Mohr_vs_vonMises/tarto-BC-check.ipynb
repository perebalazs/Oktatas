```julia
using LinearAlgebra
```

```julia
using LowLevelFEM
gmsh.initialize()

gmsh.open("tarto.geo")
```

```julia
mat = material("part")
prob = Problem([mat])
```

```julia
supp1 = displacementConstraint("base", uz=0)
supp2 = displacementConstraint("hole1", ux=0, uy=0)
supp3 = displacementConstraint("perim", uz=-0.00001)

load1 = load("hole2", fy=-5)
```

```julia
F = loadVector(prob, [load1])
showDoFResults(F, name="load")
```

```julia
resultant(F, "hole2")
```

```julia
q = solveDisplacement(prob, [load1], [supp1, supp2, supp3]);
```

```julia
showDoFResults(q, name="displacement")
```

```julia
S = solveStress(q)
showDoFResults(S)
```

```julia
b = -S ⋅ ∇;
```

```julia
ex = VectorField(prob, mat.phName, [1, 0, 0])
ey = VectorField(prob, mat.phName, [0, 1, 0])
ez = VectorField(prob, mat.phName, [0, 0, 1])

bx = b * ex
by = b * ey
bz = b * ez

load2 = load("part", fx=bx, fy=by, fz=bz);
```

```julia
q2 = solveDisplacement(prob, [load1, load2], [supp1, supp2, supp3]);
showDoFResults(q2, name="disp. mod.")
```

```julia
S2 = solveStress(q2)
showDoFResults(S2)
```

```julia
showDoFResults(S2 - S)
```
