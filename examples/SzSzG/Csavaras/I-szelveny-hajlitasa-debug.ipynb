```julia
using LowLevelFEM, LinearAlgebra
gmsh.initialize()

#gmsh.open("I-szelveny-transfinite.geo")
gmsh.open("I-szelveny-2.geo")
gmsh.model.mesh.removeDuplicateNodes()
```

```julia
mat = material("body")
prob = Problem([mat], bandwidth=:Metis)
```

```julia
supp1 = displacementConstraint("left", ux=0, uy=0, uz=0)
load1 = load("right", fy=-1)
```

```julia
K = stiffnessMatrix(prob)
```

```julia
f = loadVector(prob, [load1]);
```

```julia
#q = solveDisplacement(prob, [load1], [supp1], condensed=false);
```

```julia
fixed = constrainedDoFs(prob, [supp1]);
```

```julia
free = setdiff(1:prob.non*prob.pdim, fixed);
```

```julia
u = copy(f);
```

```julia
fill!(u.a, 0.0);
```

```julia
applyBoundaryConditions!(u, [supp1])
```

```julia
f_kin = K.A[:, fixed] * u.a[fixed];
```

```julia
K1 = K.A[free, free];
```

```julia
f1 = f.a[free] - f_kin[free];
```

```julia
K2 = Symmetric(K1)
```

```julia
A = cholesky(K2);
```

```julia
u1 = similar(f1);
```

```julia
ldiv!(u1, A, f1);
```

```julia
#u1 = K1 \ f1;
```

```julia
u.a[free] = u1;
```

```julia
#u.a[free] = K.A[free, free] \ (f.a[free] - f_kin[free]);
```

```julia
q = u;
```

```julia
showDoFResults(q)
```

```julia
S = solveStress(q, DoFResults=true);
```

```julia
n1 = normalVector(prob, "mid");
```

```julia
showDoFResults(n1)
```

```julia
S1 = nodesToElements(S, onPhysicalGroup="mid");
```

```julia
Sn = S1 * n1;
```

```julia
St = Sn - (Sn * n1) * n1;
```

```julia
showElementResults(Sn)
```

```julia
st2 = showElementResults(St)
```

```julia
st3 = showDoFResults(St)
```

```julia
plotOnPath("path", st3)
```
