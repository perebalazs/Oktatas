```julia
using LowLevelFEM
gmsh.initialize()

gmsh.open("I-szelveny.geo")
```

```julia
mat = material("body")
prob = Problem([mat])
```

```julia
supp1 = displacementConstraint("left", ux=0, uy=0, uz=0)
load1 = load("right", fy=-1)
```

```julia
f = loadVector(prob, [load1]);
```

```julia
K = stiffnessMatrix(prob);
```

```julia
applyBoundaryConditions!(K, f, [supp1])
```

```julia
q = K \ f;
```

```julia
S = solveStress(q, DoFResults=true);
```

```julia
S = solveStress(q);
```

```julia
S = elementsToNodes(S);
```

```julia
n1 = normalVector(prob, "left");
```

```julia
showDoFResults(n1)
```

```julia
S0 = nodesToElements(S);
```

```julia
S0 = nodesToElements(S, onPhysicalGroup="left");
```

```julia
n2 = elementsToNodes(n1);
```

```julia
Sn = S * n2;
```

```julia
Sn2 = S0 * n1;
```

```julia
showDoFResults(Sn)
```

```julia
showDoFResults(Sn2)
```
