```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
#gmsh.open("torus.geo")
```

```julia
r = 5
R = 10

setParameter("r", r)
setParameter("R", R)

gmsh.merge("torus.geo")
```

```julia
mat = material("volu")
prob = Problem([mat])
```

```julia
vx(x, y, z) = x * y * z / 100
vy(x, y, z) = (x^2 + y^2 + z^2) / 10
vz(x, y, z) = (2 * x * y + 2 * y * z + 2 * z * 1) / 10
```

```julia
n = normalVector(prob, "surf")
v_surf = VectorField(prob, "surf", [vx, vy, vz])
v_volu = VectorField(prob, "volu", [vx, vy, vz]);
```

```julia
integrate(prob, "surf", v_surf ⋅ n)
```

```julia
integrate(prob, "volu", ∇ ⋅ v_volu)
```

```julia
integrate(prob, "surf", (x, y, z) -> 1)
```

```julia
2 * r * π * 2 * R * π
```

```julia
@time integrate(prob, "volu", (x, y, z) -> 1)
```

```julia
r^2 * π * 2 * R * π
```

```julia
showElementResults(n, name="n")
showElementResults(v_surf, name="v surf")
showElementResults(v_volu, name="v volu")
showElementResults(v_surf ⋅ n, name="v_n")
showElementResults(div(v_volu), name="div v")
```
