```julia
using LinearAlgebra
```

```julia
n = normalVector(prob, "surf")
showElementResults(n, name="normal")
```

```julia
tx = tangentVector(prob, "x")
ty = tangentVector(prob, "y")
tz = tangentVector(prob, "z")

showElementResults(tx, name="tx")
showElementResults(ty, name="ty")
showElementResults(tz, name="tz")
```

```julia
isElementwise(tx)
```

```julia
nx = nodesToElements(elementsToNodes(grad(tx)), onPhysicalGroup="x") * tx
showElementResults(nx, name="grad(tx)*tx")
```

```julia
grad_tx = nodesToElements(elementsToNodes(grad(tx)), onPhysicalGroup="x")
ex = VectorField(prob, "x", [1, 0, 0])
ey = VectorField(prob, "x", [0, 1, 0])
ez = VectorField(prob, "x", [0, 0, 1])
showElementResults(grad_tx * ex, name="grad(tx)*ex")
showElementResults(grad_tx * ey, name="grad(tx)*ey")
showElementResults(grad_tx * ez, name="grad(tx)*ez")
```
