```julia
using LowLevelFEM, LinearAlgebra
```

```julia
gmsh.initialize()
gmsh.open("sphere-part.geo")

#openPreProcessor()
```

```julia
#mat = material("surf")
mat = material("volu")
prob = Problem([mat])
```

```julia
f(x, y, z) = 1
```

```julia
integrate(prob, "surf", f)
```

```julia
10^2 * π * 4 / 8
```

```julia
integrate(prob, "peri", f)
```

```julia
10 * π / 2 * 3
```

```julia
t = tangentVector(prob, "peri")

showElementResults(t, name="t elementwise")
```

```julia
n = LowLevelFEM.normalVector1D(prob, "peri")

showElementResults(n, name="n elementwise")
showElementResults(n / norm(n) / norm(n), name="n1 elementwise")
```
