```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("triangle.geo")
```

```julia
mat = Material("surf")
prob = Problem([mat], type=:ScalarField)
```

```julia
u = ScalarField(prob, "surf", (x, y, z) -> x == 0 && y == 0.25 ? 1 : 0)
u0 = showElementResults(u, visible=true)
gmsh.view.option.setNumber(u0, "NormalRaise", 1)
gmsh.view.option.setNumber(u0, "AdaptVisualizationGrid", 1)
gmsh.view.option.setNumber(u0, "MaxRecursionLevel", 4)
openPostProcessor()
```

```julia
gmsh.finalize()
```
