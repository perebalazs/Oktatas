```julia
using LowLevelFEM
```

```julia
gmsh.initialize()

structured_rect_mesh(lx=1)
```

```julia
mat = Material("surface")
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f);
```

```julia
supp1 = BoundaryCondition("left", ux=0)
supp2 = BoundaryCondition("bottom", uy=0)

load = LoadCondition("right", fx=1);
```

```julia
E = 1#mat.E
ОҪ = 0#mat.ОҪ

Ој = mat.Ој
О» = mat.О»

D = E / ((1 + ОҪ) * (1 - 2ОҪ)) * [1-ОҪ ОҪ 0; ОҪ 1-ОҪ 0; 0 0 (1-2ОҪ)/2]
D = E / (1 - ОҪ^2) * [1 ОҪ 0; ОҪ 1 0; 0 0 (1-ОҪ)/2]
```

```julia
K = вҲ«(Оө(Pu) вӢ… D вӢ… Оө(Pu))
K = вҲ«(Оө(Pu) вӢ… Оө(Pu) * 2Ој + Div(Pu) вӢ… Div(Pu) * О»)
K[:, :]
```

```julia
f = loadVector(Pu, [load])
```

```julia
u = solveField(K, f, support=[supp])
```

```julia
showDoFResults(u, name="u", factor=1)
```

```julia
u = expandTo3D(u)

A = (u вҲҳ вҲҮ + вҲҮ вҲҳ u) / 2
```

```julia
I = unitTensor(A)

F = E / (1 + ОҪ) * (A + ОҪ / (1 - 2ОҪ) * trace(A) * I)
```

```julia
showStressResults(F, :sx, name="Пғx")
```
